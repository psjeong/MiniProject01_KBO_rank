```python
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from datetime import datetime, timedelta
import time
import pandas as pd
import re
import zipfile
import os
import requests
from tqdm import tqdm
```

```python
url_base = 'https://www.koreabaseball.com/Record/Player/{category}'
category_list = ['HitterBasic/Basic1.aspx']
category_list_3 = ['HitterBasic/Detail1.aspx']
driver = webdriver.Chrome()
driver.implicitly_wait(3)
```

```python
for category in category_list_3:
    url = url_base.format(category=category)
driver.get(url)
driver.maximize_window()
```

```python
team_l =['','OB','LT','SS','WO','HH','HT','KT','LG','NC','SK']
```

```python
from tqdm import tqdm
baseball_data = []
for category in category_list:
    url = url_base.format(category = category)
    driver.get(url)
    driver.maximize_window()
    team_l =['OB','LT','SS','WO','HH','HT','KT','LG','NC','SK','']    
    category_data_list = []
    for team in tqdm(team_l):
        driver.find_elements_by_class_name('select02')[1].click()
        xpath_format = '//option[@value="{team}"]'
        driver.find_element_by_xpath(xpath_format.format(team = team)).click()
        
        time.sleep(0.1)
        temp_list = [x.split(' ') for x in driver.find_element_by_tag_name('table').text.split('\n')]
        temp_data = pd.DataFrame(columns = temp_list[0])
        
        for i,temp in enumerate(temp_list[1:]):
            temp_data.loc[i] = temp
            time.sleep(0.5)
        category_data_list.append(temp_data)
        time.sleep(0.1)
        time.sleep(1)
    basic_1 = pd.concat(category_data_list)
    baseball_data.append(basic_1)
```

```python
category_list_2 = ['HitterBasic/Basic2.aspx']
from tqdm import tqdm
baseball_data = []
for category in category_list_2:
    url = url_base.format(category = category)
    driver.get(url)
    driver.maximize_window()
    team_l =['OB','LT','SS','WO','HH','HT','KT','LG','NC','SK','']    
    category_data_list = []
    for team in tqdm(team_l):
        driver.find_elements_by_class_name('select02')[1].click()
        xpath_format = '//option[@value="{team}"]'
        driver.find_element_by_xpath(xpath_format.format(team = team)).click()
        
        time.sleep(0.1)
        temp_list = [x.split(' ') for x in driver.find_element_by_tag_name('table').text.split('\n')]
        temp_data = pd.DataFrame(columns = temp_list[0])
        
        for i,temp in enumerate(temp_list[1:]):
            temp_data.loc[i] = temp
            time.sleep(0.5)
        category_data_list.append(temp_data)
        time.sleep(0.1)
        time.sleep(1)
    basic_2 = pd.concat(category_data_list)
    baseball_data.append(basic_2)
```
